```python
import re
```

```python
shakey = ""
with open('shakespeare.txt') as f:
    shakey = f.read().replace('\n\n', ' ')

shakey = re.sub(r' *[0-9]+ *\n', "", shakey)

shakey = re.sub(r'\n', " ", shakey)

shakey = re.sub(r' +', " ", shakey)

shakey = re.sub(r'<<.+?>>', " ", shakey)

with open('shakespeare_cleaned.txt', 'w') as f:
    f.write(shakey)
```
